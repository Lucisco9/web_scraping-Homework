```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import Select
from fake_useragent import UserAgent
from time import sleep

my_options = webdriver.ChromeOptions()
# my_options.add_argument("--headless")                #不開啟實體瀏覽器背景執行
# my_options.add_argument("--start-maximized")         #最大化視窗
my_options.add_argument("--incognito")               #開啟無痕模式
my_options.add_argument("--disable-popup-blocking") #禁用彈出攔截
my_options.add_argument("--disable-notifications")  #取消通知

ua = UserAgent(use_external_data=True)
headers = {
    'user-agent': ua.random
}

driver = webdriver.Chrome(
    options = my_options,
    service = Service(ChromeDriverManager().install())
)
driver.get('https://irs.thsrc.com.tw/IMINT/?locale=tw&_ga=2.53668641.804746703.1682846641-1160614170.1671203198')

ac = ActionChains(driver)
#先接受cookie
wait = WebDriverWait(driver, 10)
cookie_alert = wait.until(EC.element_to_be_clickable((By.ID, "cookieAccpetBtn")))
cookie_alert.click()

#選擇單程還是來回
tripType = driver.find_element(By.ID, "BookingS1Form_tripCon_typesoftrip")
ac.move_to_element(tripType).perform()
selectType = Select(tripType).select_by_value("1") #0是單程,1是來回

#起始站南港
chooseStart = driver.find_element(By.NAME, "selectStartStation")
ac.move_to_element(chooseStart).perform()
selectStart=Select(chooseStart).select_by_value("1")

#到達站左營
chooseEnd = driver.find_element(By.NAME, "selectDestinationStation")
ac.move_to_element(chooseEnd).perform()
selectStart=Select(chooseEnd).select_by_value("12")


#出發日期
depart_date="May 12, 2023"
chooseDepart = driver.find_element(By.XPATH, f".//div[@data-triptype='oneway']")
ac.move_to_element(chooseDepart).click().perform()
departD = driver.find_element(By.XPATH, f".//span[@aria-label='{depart_date}']")
ac.move_to_element(departD).click().perform()



#出發時間
toTime='830A'
chooseToTime = driver.find_element(By.XPATH, f".//select[@name='toTimeTable']")
ac.move_to_element(chooseToTime).click().perform()
selectToTime=Select(chooseToTime).select_by_value(toTime)


# #回程日期
# return_date="May 30, 2023"
# chooseReturn = driver.find_element(By.XPATH, f".//div[@data-triptype='round-trip']")
# ac.move_to_element(chooseReturn).click().perform()
# returnD = driver.find_element(By.XPATH, f".//span[@aria-label='{return_date}']")
# ac.move_to_element(returnD).click().perform()



#回程時間
backTime='1030P'
chooseBackTime = driver.find_element(By.XPATH, f".//select[@name='backTimeTable']")
ac.move_to_element(chooseBackTime).click().perform()
selectBackTime=Select(chooseBackTime).select_by_value(backTime)
```
